```python
import pandas as pd
import numpy as np
```

```python
Data = {'S':['a','a','b','b','c', 'k','a','b','a'], 'B':['b','b','a','a','b','a','i','u','b'],
       'V':[10,7,23,54,34,5,90,8,5], 'D':[1,2,3,4,5,6,7,8,9]}
df = pd.DataFrame(Data)
df
```

```python
df1 = df.rename(columns={'S': 'B', 'B': 'S', 'V':'V1','D':'D1'}) 
df1
```

```python
df_thinned = pd.merge(df.drop_duplicates(), df1, on=['S', 'B'], how='inner')
df_thinned
```

```python
df_thinned['diff'] = df_thinned['D1'] - df_thinned['D']
df_thinned1 = df_thinned[df_thinned['diff'] > 0]
df_thinned1
```

```python
sorted_df = df_thinned1.sort_values(by=['D1'], ascending=True)
sorted_df
```

```python
s = np.unique(df_thinned1['S'])
b = np.unique(df_thinned1['B'])
```
